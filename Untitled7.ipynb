```python
import os
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model, load_model
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
import matplotlib.pyplot as plt

# Define base path - update this to match your dataset structure
BASE_DIR = r'C:\Users\Anjali Patel\Documents\archive (1)\Indian Food Images'

# Set parameters
IMG_SIZE = 224  # MobileNetV2 default input size
BATCH_SIZE = 32
EPOCHS = 10
NUM_CLASSES = 20  # Update this based on your dataset

# Load your salt content dataset
def load_salt_data():
    salt_df = pd.read_csv('salt_data.csv')
    salt_content_dict = {}
    
    for _, row in salt_df.iterrows():
        food_key = row['Food_Name'].lower().replace(' ', '_')
        salt_content_dict[food_key] = {
            'salt_content_mg': row['Salt_Content_mg_per_100g'],
            'category': row['Category'],
            'region': row['Region'],
            'main_ingredients': row['Main_Ingredients'],
            'typical_portion_size_g': row['Typical_Portion_Size_g'],
            'salt_per_typical_portion_mg': row['Salt_Per_Typical_Portion_mg']
        }
    
    return salt_content_dict

# Load the salt content data
indian_food_salt_content = load_salt_data()

# Function to create and compile the CNN model
def create_food_recognition_model():
    base_model = MobileNetV2(
        input_shape=(IMG_SIZE, IMG_SIZE, 3),
        include_top=False,
        weights='imagenet'
    )
    
    base_model.trainable = False
    
    model = Sequential([
        base_model,
        GlobalAveragePooling2D(),
        Dense(512, activation='relu'),
        Dropout(0.5),
        Dense(NUM_CLASSES, activation='softmax')
    ])
    
    model.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model

# Data augmentation for training
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Only rescale for validation/testing
val_datagen = ImageDataGenerator(rescale=1./255)

# Load and preprocess the data
def load_data():
    # Create a list of all image directories
    food_classes = os.listdir(BASE_DIR)
    food_classes = [f for f in food_classes if os.path.isdir(os.path.join(BASE_DIR, f))]
    
    # Split data into training and validation sets
    train_data = []
    val_data = []
    
    for food_class in food_classes:
        food_path = os.path.join(BASE_DIR, food_class)
        images = os.listdir(food_path)
        np.random.shuffle(images)
        
        split_index = int(len(images) * 0.8)  # 80% for training, 20% for validation
        train_data.extend([(os.path.join(food_path, img), food_class) for img in images[:split_index]])
        val_data.extend([(os.path.join(food_path, img), food_class) for img in images[split_index:]])
    
    # Create generators
    train_generator = train_datagen.flow_from_dataframe(
        pd.DataFrame(train_data, columns=['filename', 'class']),
        x_col='filename',
        y_col='class',
        target_size=(IMG_SIZE, IMG_SIZE),
        batch_size=BATCH_SIZE,
        class_mode='categorical'
    )

    validation_generator = val_datagen.flow_from_dataframe(
        pd.DataFrame(val_data, columns=['filename', 'class']),
        x_col='filename',
        y_col='class',
        target_size=(IMG_SIZE, IMG_SIZE),
        batch_size=BATCH_SIZE,
        class_mode='categorical'
    )
    
    return train_generator, validation_generator

# Create and train the model
def train_model(train_generator, validation_generator):
    model = create_food_recognition_model()
    
    checkpoint = ModelCheckpoint(
        'best_food_model.h5',
        monitor='val_accuracy',
        save_best_only=True,
        mode='max'
    )
    
    early_stopping = EarlyStopping(
        monitor='val_loss',
        patience=5,
        restore_best_weights=True
    )
    
    history = model.fit(
        train_generator,
        steps_per_epoch=train_generator.samples // BATCH_SIZE,
        epochs=EPOCHS,
        validation_data=validation_generator,
        validation_steps=validation_generator.samples // BATCH_SIZE,
        callbacks=[checkpoint, early_stopping]
    )
    
    return model, history

# Plot training history
def plot_training_history(history):
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')
    
    plt.tight_layout()
    plt.savefig('training_history.png')
    plt.show()

    
# Enhanced personalized stimulation adjustment based on user health profile and food salt content
def personalized_stimulation(salt_content, user_id=1):
    """
    Adjusts the stimulation level based on the salt content and user's health profile.
    
    Parameters:
    salt_content (float): The salt content in mg per 100g.
    user_id (int): The ID of the user for personalized stimulation.
    
    Returns:
    float: The adjusted stimulation level on a scale of 0 to 10.
    """
    # Get user health profile
    if user_id in user_health_profiles:
        user_profile = user_health_profiles[user_id]
    else:
        # Default profile if user ID not found
        user_profile = {
            'condition': 'none', 
            'severity': 'none',
            'sodium_restriction_mg': 2300,  # Default sodium restriction for healthy adults
            'taste_sensitivity': 'normal'
        }
    
    # Different logic for users with no health conditions
    if user_profile['condition'] == 'none':
        # For healthy users, stimulation is based more on taste preference than medical need
        
        # Base stimulation based on salt content for healthy users
        # Lower stimulation for high-salt foods (letting natural taste come through)
        # Higher stimulation for low-salt foods (enhancing subtle flavors)
        if salt_content > 500:  # High salt content
            base_stimulation = 2.0
        elif salt_content > 300:  # Medium salt content
            base_stimulation = 3.0
        elif salt_content > 150:  # Low salt content
            base_stimulation = 4.0
        else:  # Very low salt content
            base_stimulation = 5.0
            
        # Taste sensitivity adjustments for healthy users
        sensitivity_adjustment = {
            'highly_reduced': 2.0,      # Needs much more stimulation
            'reduced': 1.5,             # Needs more stimulation
            'slightly_reduced': 1.0,    # Needs slightly more stimulation
            'normal': 0,                # No adjustment
            'slightly_enhanced': -1.0,  # Needs less stimulation
            'enhanced': -1.5            # Needs much less stimulation
        }
        
        # Age-based adjustment (older people often have reduced taste sensitivity)
        age = user_profile['age']
        if age > 60:
            age_adjustment = 1.0
        elif age > 50:
            age_adjustment = 0.5
        elif age > 40:
            age_adjustment = 0.2
        else:
            age_adjustment = 0
            
        # Calculate final stimulation for healthy users
        final_stimulation = (
            base_stimulation + 
            sensitivity_adjustment.get(user_profile['taste_sensitivity'], 0) +
            age_adjustment
        )
        
    else:
        # Original logic for users with health conditions
        
        # Calculate base stimulation based on the ratio of salt content to user's sodium restriction
        daily_limit = user_profile['sodium_restriction_mg']
        portion_salt_percentage = (salt_content / daily_limit) * 100
        
        # Base stimulation calculation
        # Higher salt content relative to restriction = higher stimulation needed
        if portion_salt_percentage > 30:  # Very high salt content relative to restriction
            base_stimulation = 8.0
        elif portion_salt_percentage > 20:  # High salt content
            base_stimulation = 6.5
        elif portion_salt_percentage > 10:  # Moderate salt content
            base_stimulation = 5.0
        elif portion_salt_percentage > 5:   # Low salt content
            base_stimulation = 3.5
        else:  # Very low salt content
            base_stimulation = 2.0
        
        # Adjust for condition severity
        severity_adjustment = {
            'none': 0,
            'mild': 0.5,
            'moderate': 1.0,
            'severe': 1.5
        }
        
        # Adjust for taste sensitivity
        sensitivity_adjustment = {
            'highly_reduced': 1.0,     # Needs more stimulation
            'reduced': 0.5,            # Needs slightly more stimulation
            'normal': 0,               # No adjustment
            'enhanced': -0.5           # Needs less stimulation
        }
        
        # Condition-specific adjustments
        condition_adjustment = 0
        if user_profile['condition'] == 'hypertension':
            # Hypertension patients need more taste enhancement for low-salt foods
            condition_adjustment = 0.5
        elif user_profile['condition'] == 'kidney_disease':
            # Kidney disease patients often have altered taste perception
            condition_adjustment = 0.7
        elif user_profile['condition'] == 'heart_failure':
            # Heart failure patients typically need strict sodium restrictions
            condition_adjustment = 0.6
        elif user_profile['condition'] == 'liver_disease':
            # Liver disease can affect taste perception
            condition_adjustment = 0.4
        
        # Calculate final stimulation level with all adjustments
        final_stimulation = (
            base_stimulation + 
            severity_adjustment.get(user_profile['severity'], 0) +
            sensitivity_adjustment.get(user_profile['taste_sensitivity'], 0) +
            condition_adjustment
        )
    
    # Ensure the stimulation level is within the valid range (0-10)
    final_stimulation = max(0, min(10, final_stimulation))
    
    # Round to 2 decimal places for precision
    return round(final_stimulation, 2)


# Generate personalized dietary and usage recommendations
def generate_recommendations(food_name, nutrition_data, user_id=1):
    """
    Generates personalized dietary recommendations based on the food's nutritional information
    and the user's health profile.
    
    Parameters:
    food_name (str): The name of the food.
    nutrition_data (dict): A dictionary containing the food's nutritional information.
    user_id (int): The ID of the user for personalized recommendations.
    
    Returns:
    list: A list of recommendation strings.
    """
    # Get user health profile
    if user_id in user_health_profiles:
        user_profile = user_health_profiles[user_id]
    else:
        # Default profile if user ID not found
        user_profile = {
            'condition': 'none', 
            'severity': 'none',
            'sodium_restriction_mg': 2300,  # Default sodium restriction
            'taste_sensitivity': 'normal'
        }
    
    recommendations = []
    
    # Basic food information
    recommendations.append(f"{food_name} contains approximately {nutrition_data['salt_content_mg']} mg of salt per 100g.")
    
    # Typical portion information
    recommendations.append(f"A typical portion ({nutrition_data['typical_portion_size_g']}g) contains {nutrition_data['salt_per_typical_portion_mg']} mg of salt.")
    
    # Different recommendations for users with and without health conditions
    if user_profile['condition'] == 'none':
        # For users without health conditions
        daily_salt_limit = 2300  # General recommendation for healthy adults
        salt_percent_daily = (nutrition_data['salt_per_typical_portion_mg'] / daily_salt_limit) * 100
        
        recommendations.append(f"This represents approximately {salt_percent_daily:.1f}% of the general recommended daily salt intake (2300 mg).")
        
        # General healthy eating advice
        if salt_percent_daily > 30:
            recommendations.append("This dish has high salt content. Consider balancing your salt intake for the rest of the day.")
        
        # Smart Spoon recommendations based on taste preference
        if user_profile['taste_sensitivity'] == 'enhanced':
            recommendations.append("Based on your enhanced taste sensitivity, a lower stimulation level is recommended.")
        elif user_profile['taste_sensitivity'] in ['reduced', 'highly_reduced']:
            recommendations.append("Based on your reduced taste sensitivity, a higher stimulation level is recommended to enhance flavor perception.")
        
        # Add age-specific recommendations if older
        if user_profile['age'] > 60:
            recommendations.append("As taste sensitivity may naturally decrease with age, you might benefit from higher stimulation levels.")
        
    else:
        # For users with health conditions (original code)
        # Personalized daily intake percentage
        daily_salt_limit = user_profile['sodium_restriction_mg']
        salt_percent_daily = (nutrition_data['salt_per_typical_portion_mg'] / daily_salt_limit) * 100
        recommendations.append(f"This represents approximately {salt_percent_daily:.1f}% of your recommended daily salt limit ({daily_salt_limit} mg).")
        
        # Condition-specific recommendations
        if user_profile['condition'] == 'hypertension':
            if user_profile['severity'] in ['moderate', 'severe']:
                if salt_percent_daily > 15:
                    recommendations.append("CAUTION: This dish exceeds the recommended salt intake for your hypertension condition.")
                    recommendations.append("Consider reducing portion size or choosing a lower-sodium alternative.")
                else:
                    recommendations.append("This dish is within acceptable sodium limits for your condition.")
        
        elif user_profile['condition'] == 'kidney_disease':
            if salt_percent_daily > 10:
                recommendations.append("CAUTION: This dish may not be suitable for your kidney condition.")
                recommendations.append("Consider speaking with your dietitian before consuming this dish.")
            else:
                recommendations.append("This dish is within acceptable sodium limits for your condition.")
        
        elif user_profile['condition'] == 'heart_failure':
            if salt_percent_daily > 12:
                recommendations.append("CAUTION: This dish exceeds the recommended salt intake for heart failure patients.")
                recommendations.append("Consider reducing portion size or choosing a lower-sodium alternative.")
            else:
                recommendations.append("This dish is within acceptable sodium limits for your condition.")
        
        # Taste sensitivity recommendations
        if user_profile['taste_sensitivity'] == 'reduced' or user_profile['taste_sensitivity'] == 'highly_reduced':
            recommendations.append("Your taste sensitivity profile suggests you may benefit from enhanced flavor stimulation.")
    
    # General recommendations for everyone
    if nutrition_data['salt_content_mg'] > 500:
        recommendations.append("Consider pairing this dish with fresh vegetables or a simple salad to balance salt intake.")
    
    # Smart Spoon usage recommendation
    recommendations.append(f"Recommended Smart Spoon stimulation level: {personalized_stimulation(nutrition_data['salt_content_mg'], user_id):.2f}/10")
    
    return recommendations


import matplotlib.pyplot as plt

def visualize_salt_content(salt_content, food_name):
    """
    Visualizes the salt content of a food item.
    
    Parameters:
    salt_content (float): The salt content in mg per 100g.
    food_name (str): The name of the food item.
    """
    # Define the salt content thresholds for different visualization levels
    low_threshold = 120
    medium_threshold = 480
    high_threshold = 800
    
    # Create a figure and axis
    fig, ax = plt.subplots(figsize=(8, 6))
    
    # Plot the salt content bar
    ax.bar(0, salt_content, color='blue')
    
    # Set the x-axis labels and title
    ax.set_xticks([0])
    ax.set_xticklabels([food_name])
    ax.set_title(f"Salt Content of {food_name}")
    
    # Set the y-axis label
    ax.set_ylabel("Salt Content (mg per 100g)")
    
    # Add horizontal lines for the salt content thresholds
    ax.axhline(y=low_threshold, color='green', linestyle='--', label='Low')
    ax.axhline(y=medium_threshold, color='yellow', linestyle='--', label='Medium')
    ax.axhline(y=high_threshold, color='red', linestyle='--', label='High')
    
    # Add a legend
    ax.legend()
    
    # Display the plot
    plt.show()
    
    
# (The rest of your functions remain unchanged)
def get_food_nutrition_data(predicted_class, class_indices):
    # Reverse the class indices to get food name from the class index
    food_name = list(class_indices.keys())[predicted_class]
    
    # Convert food name to the format used in salt content dictionary
    food_key = food_name.lower().replace(' ', '_')
    
    # Retrieve the nutrition data from the salt content dictionary
    if food_key in indian_food_salt_content:
        nutrition_data = indian_food_salt_content[food_key]
        return nutrition_data, food_name
    else:
        raise ValueError(f"Nutrition data for {food_name} not found.")

# Main prediction function
def predict_and_process_image(model, image_path, class_indices,user_id):
    img = tf.keras.preprocessing.image.load_img(image_path, target_size=(IMG_SIZE, IMG_SIZE))
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0) / 255.0
    
    predictions = model.predict(img_array)
    predicted_class = np.argmax(predictions[0])
    confidence = predictions[0][predicted_class]
    
    nutrition_data, food_name = get_food_nutrition_data(predicted_class, class_indices)
    salt_content = nutrition_data['salt_content_mg']
    stimulation = personalized_stimulation(salt_content, user_id)
    recommendations = generate_recommendations(food_name, nutrition_data,user_id)
    
    result = {
        'food_name': food_name.replace('_', ' ').title(),
        'confidence': float(confidence),
        'salt_content_mg': salt_content,
        'salt_per_portion_mg': nutrition_data['salt_per_typical_portion_mg'],
        'typical_portion_size_g': nutrition_data['typical_portion_size_g'],
        'category': nutrition_data['category'],
        'region': nutrition_data['region'],
        'main_ingredients': nutrition_data['main_ingredients'],
        'stimulation_level': stimulation,
        'recommendations': recommendations
    }
    
    return result
```

```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
import torch
from wordcloud import WordCloud
import warnings
warnings.filterwarnings('ignore')

# Download necessary NLTK resources
nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)

class SmartSpoonFeedbackAnalyzer:
    def __init__(self, data_path, test_size=0.2, random_state=42):
        """Initialize the analyzer with dataset and configuration parameters."""
        self.data_path = data_path
        self.test_size = test_size
        self.random_state = random_state
        self.data = None
        self.X_train = None
        self.X_test = None
        self.y_train = None
        self.y_test = None
        self.sentiment_model = None
        self.behavior_model = None
        self.label_encoder = LabelEncoder()
        self.stopwords = set(stopwords.words('english'))
        self.lemmatizer = WordNetLemmatizer()
        
        # Load pre-trained sentiment analysis model
        try:
            self.sentiment_pipeline = pipeline(
                "sentiment-analysis",
                model="distilbert-base-uncased-finetuned-sst-2-english",
                tokenizer="distilbert-base-uncased-finetuned-sst-2-english"
            )
        except:
            print("Warning: Failed to load transformers model. Falling back to simpler methods.")
            self.sentiment_pipeline = None

    def load_data(self):
        """Load and preprocess the dataset."""
        try:
            # Attempt different encodings
            encodings = ['utf-8', 'ISO-8859-1', 'cp1252', 'utf-16']
            for encoding in encodings:
                try:
                    self.data = pd.read_csv(self.data_path, encoding=encoding)
                    break
                except UnicodeDecodeError:
                    continue
            
            if self.data is None:
                raise ValueError("Could not decode file with any of the attempted encodings")
                
            # Check for required columns
            required_columns = ['Feedback_Text', 'Satisfaction_Level']
            for col in required_columns:
                if col not in self.data.columns:
                    raise ValueError(f"Required column '{col}' not found in the dataset")
            
            # Handle missing values
            self.data['Feedback_Text'].fillna("", inplace=True)
            
            # Remove rows with missing satisfaction levels
            self.data = self.data.dropna(subset=['Satisfaction_Level'])
            
            # Convert satisfaction levels to string for classification
            self.data['Satisfaction_Level'] = self.data['Satisfaction_Level'].astype(str)
            
            print(f"Data loaded successfully with {self.data.shape[0]} rows and {self.data.shape[1]} columns")
            return True
            
        except Exception as e:
            print(f"Error loading data: {e}")
            return False

    def preprocess_text(self, text):
        """Clean and preprocess text data."""
        if not isinstance(text, str) or text == "":
            return ""
        
        # Convert to lowercase
        text = text.lower()
        
        # Remove special characters and numbers
        text = re.sub(r'[^a-zA-Z\s]', '', text)
        
        # Tokenize
        tokens = word_tokenize(text)
        
        # Remove stopwords and lemmatize
        clean_tokens = [self.lemmatizer.lemmatize(token) for token in tokens if token not in self.stopwords]
        
        return ' '.join(clean_tokens)

    def analyze_sentiment_with_transformers(self, text):
        """Analyze sentiment using pre-trained transformer model."""
        if not self.sentiment_pipeline or not isinstance(text, str) or text.strip() == "":
            return {"label": "NEUTRAL", "score": 0.5}
        
        try:
            result = self.sentiment_pipeline(text)[0]
            return result
        except Exception as e:
            print(f"Error in transformer sentiment analysis: {e}")
            return {"label": "NEUTRAL", "score": 0.5}

    def perform_sentiment_analysis(self):
        """Perform sentiment analysis on the dataset."""
        if self.data is None:
            print("Data not loaded. Please load data first.")
            return
        
        # Apply preprocessing to feedback text
        self.data['Processed_Text'] = self.data['Feedback_Text'].apply(self.preprocess_text)
        
        # Perform sentiment analysis
        print("Performing sentiment analysis...")
        
        if self.sentiment_pipeline:
            # Use transformer-based model
            sentiments = self.data['Feedback_Text'].apply(self.analyze_sentiment_with_transformers)
            self.data['Sentiment_Label'] = sentiments.apply(lambda x: x['label'])
            self.data['Sentiment_Score'] = sentiments.apply(lambda x: x['score'] if x['label'] == 'POSITIVE' else 1 - x['score'])
            
            # Map 'POSITIVE'/'NEGATIVE' to 'Positive'/'Negative' for consistency
            label_mapping = {'POSITIVE': 'Positive', 'NEGATIVE': 'Negative'}
            self.data['Sentiment'] = self.data['Sentiment_Label'].map(lambda x: label_mapping.get(x, 'Neutral'))
            
        else:
            # Fall back to NLTK's VADER
            from nltk.sentiment import SentimentIntensityAnalyzer
            nltk.download('vader_lexicon', quiet=True)
            sia = SentimentIntensityAnalyzer()
            
            self.data['Sentiment_Score'] = self.data['Feedback_Text'].apply(
                lambda x: sia.polarity_scores(x)['compound'] if isinstance(x, str) else 0
            )
            
            # Classify sentiment based on the score
            self.data['Sentiment'] = self.data['Sentiment_Score'].apply(
                lambda x: 'Positive' if x > 0.05 else ('Negative' if x < -0.05 else 'Neutral')
            )
        
        # Count sentiment distribution
        sentiment_counts = self.data['Sentiment'].value_counts()
        print("\nSentiment Distribution:")
        for sentiment, count in sentiment_counts.items():
            print(f"{sentiment}: {count} ({count/len(self.data)*100:.1f}%)")
            
        return self.data['Sentiment'].value_counts()

    def prepare_data_for_modeling(self):
        """Prepare data for user behavior prediction modeling."""
        if self.data is None or 'Processed_Text' not in self.data.columns:
            print("Data not properly processed. Please run preprocessing first.")
            return False
        
        try:
            # Split data into features and target
            X = self.data['Processed_Text']
            y = self.data['Satisfaction_Level']
            
            # Encode the target variable
            y_encoded = self.label_encoder.fit_transform(y)
            
            # Split into training and testing sets
            self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(
                X, y_encoded, test_size=self.test_size, random_state=self.random_state, stratify=y_encoded
            )
            
            print(f"Data prepared for modeling: {len(self.X_train)} training samples, {len(self.X_test)} testing samples")
            return True
            
        except Exception as e:
            print(f"Error preparing data for modeling: {e}")
            return False

    def train_behavior_model(self):
        """Train a behavior prediction model."""
        if self.X_train is None or self.y_train is None:
            print("Training data not prepared. Please prepare data first.")
            return None
            
        try:
            print("Training user behavior prediction model...")
            
            # Create a TF-IDF vectorizer + XGBoost pipeline
            pipeline = Pipeline([
                ('tfidf', TfidfVectorizer(max_features=5000, ngram_range=(1, 2))),
                ('classifier', xgb.XGBClassifier(
                    objective='multi:softprob',
                    n_estimators=100,
                    max_depth=5,
                    learning_rate=0.1,
                    subsample=0.8,
                    colsample_bytree=0.8,
                    random_state=self.random_state,
                    use_label_encoder=False,
                    eval_metric='mlogloss'
                ))
            ])
            
            # Train the model
            self.behavior_model = pipeline.fit(self.X_train, self.y_train)
            
            # Evaluate the model
            y_pred = self.behavior_model.predict(self.X_test)
            accuracy = accuracy_score(self.y_test, y_pred)
            
            print(f"Model training complete. Test accuracy: {accuracy:.4f}")
            
            # Print classification report
            target_names = [str(self.label_encoder.inverse_transform([i])[0]) for i in range(len(self.label_encoder.classes_))]
            print("\nClassification Report:")
            print(classification_report(self.y_test, y_pred, target_names=target_names))
            
            return self.behavior_model
            
        except Exception as e:
            print(f"Error training behavior model: {e}")
            return None

    def identify_keywords(self, n_keywords=10):
        """Extract key terms influencing user satisfaction."""
        if self.behavior_model is None:
            print("Behavior model not trained. Please train the model first.")
            return {}
            
        try:
            # Get TF-IDF vectorizer from pipeline
            tfidf = self.behavior_model.named_steps['tfidf']
            classifier = self.behavior_model.named_steps['classifier']
            
            # Get feature names
            feature_names = tfidf.get_feature_names_out()
            
            # Extract feature importances from classifier
            importances = classifier.feature_importances_
            
            # Sort importances
            indices = np.argsort(importances)[::-1]
            
            # Get top keywords for each satisfaction level
            keywords_by_class = {}
            
            # For XGBoost with multiple classes
            class_keywords = {}
            for i, class_label in enumerate(self.label_encoder.classes_):
                # Get feature importance for this class
                importance_for_class = importances
                indices = np.argsort(importance_for_class)[::-1]
                
                # Get top keywords
                top_keywords = [(feature_names[idx], importance_for_class[idx]) for idx in indices[:n_keywords]]
                class_keywords[class_label] = top_keywords
            
            return class_keywords
            
        except Exception as e:
            print(f"Error identifying keywords: {e}")
            return {}

    def generate_improvement_suggestions(self, feedback_text):
        """Generate improvement suggestions based on user feedback."""
        # Basic cleaning of input
        if not isinstance(feedback_text, str) or feedback_text.strip() == "":
            return {
                'Sentiment': 'Neutral',
                'Predicted_Satisfaction': 'Unknown',
                'Improvement_Suggestions': ["Please provide more detailed feedback to receive meaningful suggestions."]
            }
            
        try:
            # Preprocess the feedback
            processed_text = self.preprocess_text(feedback_text)
            
            # Analyze sentiment
            if self.sentiment_pipeline:
                sentiment_result = self.analyze_sentiment_with_transformers(feedback_text)
                sentiment = 'Positive' if sentiment_result['label'] == 'POSITIVE' else 'Negative'
                sentiment_score = sentiment_result['score']
            else:
                from nltk.sentiment import SentimentIntensityAnalyzer
                sia = SentimentIntensityAnalyzer()
                sentiment_score = sia.polarity_scores(feedback_text)['compound']
                sentiment = 'Positive' if sentiment_score > 0.05 else ('Negative' if sentiment_score < -0.05 else 'Neutral')
            
            # Predict satisfaction level
            if self.behavior_model:
                predicted_level_encoded = self.behavior_model.predict([processed_text])[0]
                predicted_satisfaction = self.label_encoder.inverse_transform([predicted_level_encoded])[0]
            else:
                predicted_satisfaction = 'Unknown'
            
            # Generate improvement suggestions
            improvement_suggestions = []
            
            # Define keyword-based suggestion mappings (enhanced from original)
            suggestions_map = {
                'battery': [
                    "Consider increasing the battery capacity for longer usage.",
                    "Implement a low-power mode to extend battery life.",
                    "Add battery status indicators to improve user awareness."
                ],
                'grip': [
                    "Enhance the ergonomic design for a better grip.",
                    "Use non-slip materials to improve handling.",
                    "Consider adjustable grip sizes for different users."
                ],
                'customize': [
                    "Add more customizable settings tailored to user preferences.",
                    "Introduce presets for different types of food.",
                    "Develop a personalization system that learns from usage patterns."
                ],
                'customizable': [
                    "Add more customizable settings tailored to user preferences.",
                    "Introduce presets for different types of food.",
                    "Develop a personalization system that learns from usage patterns."
                ],
                'personalize': [
                    "Add more customizable settings tailored to user preferences.",
                    "Introduce presets for different types of food.",
                    "Develop a personalization system that learns from usage patterns."
                ],
                'quality': [
                    "Maintain the high-quality standards in materials and design.",
                    "Conduct regular quality checks to ensure product reliability.",
                    "Consider premium materials that enhance durability and user experience."
                ],
                'features': [
                    "Explore user-requested features like temperature control or smart alerts.",
                    "Consider integrating a companion app for enhanced functionality.",
                    "Implement AI-driven features that adapt to user behavior patterns."
                ],
                'app': [
                    "Improve the companion app interface for better usability.",
                    "Add more visualization features to track nutrition data.",
                    "Implement social sharing capabilities for recipes and food experiences."
                ],
                'color': [
                    "Introduce a variety of color options for personalization.",
                    "Offer limited edition colors based on user feedback.",
                    "Consider customizable RGB lighting elements for visual feedback."
                ],
                'nutrition': [
                    "Implement tracking features for nutritional intake.",
                    "Consider adding a recipe suggestion feature based on dietary needs.",
                    "Enhance accuracy of nutritional analysis algorithms."
                ],
                'price': [
                    "Evaluate pricing strategies to ensure competitiveness.",
                    "Consider offering bundle deals for multiple purchases.",
                    "Introduce a tiered product line with different price points."
                ],
                'salt': [
                    "Adjust the salt detection sensitivity based on user feedback.",
                    "Consider offering options for low-sodium alternatives.",
                    "Implement a feature for users to customize their preferred salt level detection."
                ],
                'temperature': [
                    "Add temperature sensing capabilities to provide food safety information.",
                    "Implement temperature-based cooking recommendations.",
                    "Consider alerts for foods that are too hot to consume safely."
                ],
                'weight': [
                    "Improve the accuracy of the weight sensing mechanism.",
                    "Add portion control features based on weight detection.",
                    "Implement personalized portion recommendations based on dietary goals."
                ],
                'connect': [
                    "Enhance Bluetooth/WiFi connectivity for more reliable data transfer.",
                    "Add seamless integration with other smart kitchen devices.",
                    "Improve cloud synchronization for multi-device users."
                ],
                'charge': [
                    "Implement wireless charging capabilities for convenience.",
                    "Improve charging efficiency to reduce downtime.",
                    "Add quick-charge functionality for users on the go."
                ],
                'clean': [
                    "Enhance water-resistant properties for easier cleaning.",
                    "Consider self-cleaning features for improved hygiene.",
                    "Redesign components for dishwasher-safe capabilities."
                ],
                'comfort': [
                    "Optimize weight distribution for better comfort during extended use.",
                    "Use softer materials at contact points for improved handling.",
                    "Consider ergonomic improvements based on diverse user testing."
                ],
                'data': [
                    "Improve data visualization for easier interpretation of nutritional information.",
                    "Implement enhanced data privacy controls for user information.",
                    "Add export functionality for integration with other health apps."
                ]
            }
            
            # Context-based suggestions based on sentiment and predicted satisfaction
            context_based_suggestions = {
                'Negative': {
                    'low': [
                        "Focus on core functionality improvements for better user satisfaction.",
                        "Consider a user experience redesign based on customer feedback.",
                        "Implement quality control measures to address potential issues."
                    ],
                    'medium': [
                        "Address specific user concerns while maintaining existing positive features.",
                        "Consider incremental improvements to gradually enhance user experience.",
                        "Develop a customer feedback program to better understand pain points."
                    ],
                    'high': [
                        "Though satisfaction is high, address negative feedback to maintain loyalty.",
                        "Consider minor refinements to address specific concerns.",
                        "Implement a proactive communication strategy to address potential issues."
                    ]
                },
                'Neutral': {
                    'low': [
                        "Focus on creating more emotional connection with the product experience.",
                        "Consider adding standout features to differentiate from competitors.",
                        "Implement user engagement strategies to increase product attachment."
                    ],
                    'medium': [
                        "Enhance unique selling points to convert neutral users to advocates.",
                        "Consider personalization features to increase user connection.",
                        "Develop community features to enhance product engagement."
                    ],
                    'high': [
                        "Build on existing strengths while adding innovative features.",
                        "Consider loyalty programs to maintain high satisfaction levels.",
                        "Implement user story collection to identify growth opportunities."
                    ]
                },
                'Positive': {
                    'low': [
                        "Focus on consistency to ensure all features meet positive expectations.",
                        "Address potential barriers to full product satisfaction.",
                        "Implement continuous improvement processes based on positive feedback."
                    ],
                    'medium': [
                        "Enhance features that receive the most positive feedback.",
                        "Consider expanding product line with complementary offerings.",
                        "Develop ambassador programs to leverage satisfied customers."
                    ],
                    'high': [
                        "Maintain high quality standards while exploring innovation opportunities.",
                        "Consider premium product variants for highly satisfied customers.",
                        "Implement case study collection to showcase positive outcomes."
                    ]
                }
            }
            
            # First add keyword-based suggestions
            for keyword, suggestions in suggestions_map.items():
                if keyword in feedback_text.lower():
                    for suggestion in suggestions:
                        if suggestion not in improvement_suggestions:
                            improvement_suggestions.append(suggestion)
            
            # Then add context-based suggestions
            # Determine satisfaction level category
            satisfaction_category = 'medium'  # default
            if predicted_satisfaction != 'Unknown':
                try:
                    satisfaction_value = float(predicted_satisfaction)
                    if satisfaction_value < 3:
                        satisfaction_category = 'low'
                    elif satisfaction_value > 4:
                        satisfaction_category = 'high'
                    else:
                        satisfaction_category = 'medium'
                except:
                    # If predicted_satisfaction is not numeric
                    if 'low' in predicted_satisfaction.lower() or '1' in predicted_satisfaction or '2' in predicted_satisfaction:
                        satisfaction_category = 'low'
                    elif 'high' in predicted_satisfaction.lower() or '4' in predicted_satisfaction or '5' in predicted_satisfaction:
                        satisfaction_category = 'high'
            
            # Add context-based suggestions
            if sentiment in context_based_suggestions and satisfaction_category in context_based_suggestions[sentiment]:
                context_suggestions = context_based_suggestions[sentiment][satisfaction_category]
                for suggestion in context_suggestions:
                    if suggestion not in improvement_suggestions:
                        improvement_suggestions.append(suggestion)
            
            # If no suggestions were found, add a default one
            if not improvement_suggestions:
                improvement_suggestions = [
                    "Thank you for your feedback! We continually monitor user experiences to improve our products.",
                    "Consider participating in our user research program to help shape future features.",
                    "We appreciate your input and will use it to enhance the Smart Spoon experience."
                ]
            
            # Limit to top 5 most relevant suggestions
            improvement_suggestions = improvement_suggestions[:5]
            
            return {
                'Sentiment': sentiment,
                'Sentiment_Score': sentiment_score,
                'Predicted_Satisfaction': predicted_satisfaction,
                'Improvement_Suggestions': improvement_suggestions
            }
            
        except Exception as e:
            print(f"Error generating improvement suggestions: {e}")
            return {
                'Sentiment': 'Error',
                'Predicted_Satisfaction': 'Error',
                'Improvement_Suggestions': ["An error occurred while processing your feedback. Please try again."]
            }

    def visualize_sentiment_distribution(self):
        """Visualize the sentiment distribution in the dataset."""
        if self.data is None or 'Sentiment' not in self.data.columns:
            print("Sentiment data not available. Please run sentiment analysis first.")
            return
            
        plt.figure(figsize=(10, 6))
        sns.countplot(x='Sentiment', data=self.data, palette='viridis')
        plt.title('Sentiment Distribution in User Feedback')
        plt.xlabel('Sentiment')
        plt.ylabel('Count')
        plt.tight_layout()
        plt.show()
    
    def visualize_satisfaction_vs_sentiment(self):
        """Visualize relationship between satisfaction levels and sentiment."""
        if self.data is None or 'Sentiment' not in self.data.columns or 'Satisfaction_Level' not in self.data.columns:
            print("Required data not available. Please run preprocessing first.")
            return
            
        plt.figure(figsize=(12, 8))
        
        # Convert satisfaction level to numeric if possible
        try:
            self.data['Satisfaction_Numeric'] = pd.to_numeric(self.data['Satisfaction_Level'], errors='coerce')
            
            # Create crosstab
            cross_tab = pd.crosstab(self.data['Satisfaction_Numeric'], self.data['Sentiment'])
            
            # Plot heatmap
            sns.heatmap(cross_tab, annot=True, cmap='YlGnBu', fmt='d', cbar=True)
            plt.title('Relationship Between Satisfaction Level and Sentiment')
            plt.xlabel('Sentiment')
            plt.ylabel('Satisfaction Level')
            
        except:
            # Alternative visualization if satisfaction levels are not numeric
            cross_tab = pd.crosstab(self.data['Satisfaction_Level'], self.data['Sentiment'])
            sns.heatmap(cross_tab, annot=True, cmap='YlGnBu', fmt='d', cbar=True)
            plt.title('Relationship Between Satisfaction Level and Sentiment')
            plt.xlabel('Sentiment')
            plt.ylabel('Satisfaction Level')
            
        plt.tight_layout()
        plt.show()
    
    def generate_word_clouds(self):
        """Generate word clouds for different sentiment categories."""
        if self.data is None or 'Sentiment' not in self.data.columns or 'Processed_Text' not in self.data.columns:
            print("Required data not available. Please run preprocessing first.")
            return
            
        sentiment_categories = self.data['Sentiment'].unique()
        
        fig, axes = plt.subplots(1, len(sentiment_categories), figsize=(20, 6))
        
        if len(sentiment_categories) == 1:
            axes = [axes]
            
        for i, sentiment in enumerate(sentiment_categories):
            # Filter text by sentiment
            text = ' '.join(self.data[self.data['Sentiment'] == sentiment]['Processed_Text'])
            
            if not text.strip():
                axes[i].text(0.5, 0.5, f"No data for {sentiment}", ha='center', va='center')
                axes[i].axis('off')
                continue
                
            # Generate word cloud
            wordcloud = WordCloud(
                width=800, height=400,
                background_color='white',
                max_words=100,
                contour_width=3,
                contour_color='steelblue'
            ).generate(text)
            
            # Display
            axes[i].imshow(wordcloud, interpolation='bilinear')
            axes[i].set_title(f'Word Cloud for {sentiment} Feedback')
            axes[i].axis('off')
            
        plt.tight_layout()
        plt.show()
    
    def run_full_analysis(self):
        """Run the complete analysis pipeline."""
        print("Starting Smart Spoon Feedback Analysis...")
        
        # Step 1: Load data
        if not self.load_data():
            return "Analysis failed during data loading."
            
        # Step 2: Perform sentiment analysis
        self.perform_sentiment_analysis()
        
        # Step 3: Prepare data for modeling
        if not self.prepare_data_for_modeling():
            return "Analysis failed during data preparation."
            
        # Step 4: Train behavior model
        if not self.train_behavior_model():
            return "Analysis failed during model training."
            
        # Step 5: Identify keywords
        key_terms = self.identify_keywords(n_keywords=15)
        
        print("\nAnalysis complete!")
        print("The system is now ready to analyze new feedback and generate personalized improvement suggestions.")
        
        return "Analysis completed successfully."

def analyze_user_feedback(feedback_text, analyzer):
    """Analyze a single user feedback and return results."""
    if analyzer is None:
        return {
            'Sentiment': 'Error',
            'Predicted_Satisfaction': 'Error',
            'Improvement_Suggestions': ["Analyzer not initialized. Please run the full analysis first."]
        }
        
    results = analyzer.generate_improvement_suggestions(feedback_text)
    
    # Print results
    print("\nAnalysis Results:")
    print(f"Sentiment: {results['Sentiment']} (Score: {results['Sentiment_Score']:.2f})")
    print(f"Predicted Satisfaction Level: {results['Predicted_Satisfaction']}")
    print("Improvement Suggestions:")
    for i, suggestion in enumerate(results['Improvement_Suggestions'], 1):
        print(f"{i}. {suggestion}")
        
    return results

# Main execution logic
if __name__ == "__main__":
    try:
        # Initialize analyzer
        analyzer = SmartSpoonFeedbackAnalyzer('response_data.csv')
        
        # Run full analysis
        analyzer.run_full_analysis()
        
        # Visualizations
        analyzer.visualize_sentiment_distribution()
        analyzer.visualize_satisfaction_vs_sentiment()
        analyzer.generate_word_clouds()
        
        # Interactive user input loop
        while True:
            user_review = input("\nPlease enter your feedback about the Smart Spoon (or type 'exit' to quit): ")
            
            if user_review.lower() == 'exit':
                print("Thank you for using the Smart Spoon Feedback Analyzer!")
                break
                
            analyze_user_feedback(user_review, analyzer)
            
    except KeyboardInterrupt:
        print("\nAnalysis terminated by user.")
    except Exception as e:
        print(f"\nAn unexpected error occurred: {e}")
```

```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
import warnings
warnings.filterwarnings('ignore')

# Define simple stopwords list to avoid NLTK dependency
STOPWORDS = {'i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', "you're", "you've",
             "you'll", "you'd", 'your', 'yours', 'yourself', 'yourselves', 'he', 'him', 'his', 
             'himself', 'she', "she's", 'her', 'hers', 'herself', 'it', "it's", 'its', 'itself', 
             'they', 'them', 'their', 'theirs', 'themselves', 'what', 'which', 'who', 'whom', 
             'this', 'that', "that'll", 'these', 'those', 'am', 'is', 'are', 'was', 'were', 'be', 
             'been', 'being', 'have', 'has', 'had', 'having', 'do', 'does', 'did', 'doing', 'a', 
             'an', 'the', 'and', 'but', 'if', 'or', 'because', 'as', 'until', 'while', 'of', 'at', 
             'by', 'for', 'with', 'about', 'against', 'between', 'into', 'through', 'during', 
             'before', 'after', 'above', 'below', 'to', 'from', 'up', 'down', 'in', 'out', 'on', 
             'off', 'over', 'under', 'again', 'further', 'then', 'once', 'here', 'there', 'when', 
             'where', 'why', 'how', 'all', 'any', 'both', 'each', 'few', 'more', 'most', 'other', 
             'some', 'such', 'no', 'nor', 'not', 'only', 'own', 'same', 'so', 'than', 'too', 'very', 
             's', 't', 'can', 'will', 'just', 'don', "don't", 'should', "should've", 'now', 'd', 
             'll', 'm', 'o', 're', 've', 'y', 'ain', 'aren', "aren't", 'couldn', "couldn't", 'didn', 
             "didn't", 'doesn', "doesn't", 'hadn', "hadn't", 'hasn', "hasn't", 'haven', "haven't", 
             'isn', "isn't", 'ma', 'mightn', "mightn't", 'mustn', "mustn't", 'needn', "needn't", 
             'shan', "shan't", 'shouldn', "shouldn't", 'wasn', "wasn't", 'weren', "weren't", 'won', 
             "won't", 'wouldn', "wouldn't"}

class SmartSpoonFeedbackAnalyzer:
    def __init__(self, data_path, test_size=0.2, random_state=42):
        """Initialize the analyzer with dataset and configuration parameters."""
        self.data_path = data_path
        self.test_size = test_size
        self.random_state = random_state
        self.data = None
        self.X_train = None
        self.X_test = None
        self.y_train = None
        self.y_test = None
        self.sentiment_model = None
        self.behavior_model = None
        self.label_encoder = LabelEncoder()
        self.stopwords = STOPWORDS  # Use our predefined stopwords

    def load_data(self):
        """Load and preprocess the dataset."""
        try:
            # Attempt different encodings
            encodings = ['utf-8', 'ISO-8859-1', 'cp1252', 'utf-16']
            for encoding in encodings:
                try:
                    self.data = pd.read_csv(self.data_path, encoding=encoding)
                    break
                except UnicodeDecodeError:
                    continue
            
            if self.data is None:
                raise ValueError("Could not decode file with any of the attempted encodings")
                
            # Check for required columns
            required_columns = ['Feedback_Text', 'Satisfaction_Level']
            for col in required_columns:
                if col not in self.data.columns:
                    raise ValueError(f"Required column '{col}' not found in the dataset")
            
            # Handle missing values
            self.data['Feedback_Text'].fillna("", inplace=True)
            
            # Remove rows with missing satisfaction levels
            self.data = self.data.dropna(subset=['Satisfaction_Level'])
            
            # Convert satisfaction levels to string for classification
            self.data['Satisfaction_Level'] = self.data['Satisfaction_Level'].astype(str)
            
            print(f"Data loaded successfully with {self.data.shape[0]} rows and {self.data.shape[1]} columns")
            return True
            
        except Exception as e:
            print(f"Error loading data: {e}")
            return False

    def preprocess_text(self, text):
        """Clean and preprocess text data."""
        if not isinstance(text, str) or text == "":
            return ""
        
        # Convert to lowercase
        text = text.lower()
        
        # Remove special characters and numbers
        text = re.sub(r'[^a-zA-Z\s]', '', text)
        
        # Simple tokenization by splitting on whitespace
        tokens = text.split()
        
        # Remove stopwords
        clean_tokens = [token for token in tokens if token not in self.stopwords]
        
        return ' '.join(clean_tokens)

    def analyze_sentiment_with_lexicon(self, text):
        """Analyze sentiment using a simple lexicon-based approach."""
        if not isinstance(text, str) or text.strip() == "":
            return {"label": "Neutral", "score": 0.5}
        
        # Simple lexicon-based sentiment analysis
        positive_words = {
            'good', 'great', 'excellent', 'amazing', 'wonderful', 'fantastic', 'outstanding',
            'best', 'perfect', 'awesome', 'brilliant', 'love', 'enjoy', 'happy', 'pleased',
            'satisfied', 'recommend', 'impressive', 'superb', 'exceptional', 'delightful',
            'convenient', 'easy', 'efficient', 'quality', 'durable', 'reliable', 'comfortable',
            'innovative', 'intuitive', 'worth', 'useful', 'helpful'
        }
        
        negative_words = {
            'bad', 'poor', 'terrible', 'awful', 'horrible', 'worst', 'disappointing', 'mediocre',
            'useless', 'waste', 'difficult', 'hard', 'complicated', 'confusing', 'annoying',
            'frustrating', 'fail', 'failure', 'problem', 'issue', 'defect', 'break', 'broke',
            'expensive', 'overpriced', 'cheap', 'flimsy', 'inconvenient', 'slow', 'unreliable',
            'uncomfortable', 'disappointed', 'hate', 'dislike', 'regret', 'avoid'
        }
        
        # Normalize text and split into words
        words = text.lower().split()
        
        # Count positive and negative words
        positive_count = sum(1 for word in words if word in positive_words)
        negative_count = sum(1 for word in words if word in negative_words)
        
        # Calculate sentiment score
        total_count = positive_count + negative_count
        if total_count == 0:
            return {"label": "Neutral", "score": 0.5}
        
        score = positive_count / total_count if total_count > 0 else 0.5
        
        # Determine sentiment label
        if score > 0.6:
            label = "Positive"
        elif score < 0.4:
            label = "Negative"
        else:
            label = "Neutral"
            
        return {"label": label, "score": score}

    def perform_sentiment_analysis(self):
        """Perform sentiment analysis on the dataset."""
        if self.data is None:
            print("Data not loaded. Please load data first.")
            return
        
        # Apply preprocessing to feedback text
        self.data['Processed_Text'] = self.data['Feedback_Text'].apply(self.preprocess_text)
        
        # Perform sentiment analysis
        print("Performing sentiment analysis...")
        
        # Use our simple lexicon-based sentiment analysis
        sentiments = self.data['Feedback_Text'].apply(self.analyze_sentiment_with_lexicon)
        self.data['Sentiment_Label'] = sentiments.apply(lambda x: x['label'])
        self.data['Sentiment_Score'] = sentiments.apply(lambda x: x['score'])
        self.data['Sentiment'] = self.data['Sentiment_Label']
        
        # Count sentiment distribution
        sentiment_counts = self.data['Sentiment'].value_counts()
        print("\nSentiment Distribution:")
        for sentiment, count in sentiment_counts.items():
            print(f"{sentiment}: {count} ({count/len(self.data)*100:.1f}%)")
            
        return self.data['Sentiment'].value_counts()

    def prepare_data_for_modeling(self):
        """Prepare data for user behavior prediction modeling."""
        if self.data is None or 'Processed_Text' not in self.data.columns:
            print("Data not properly processed. Please run preprocessing first.")
            return False
        
        try:
            # Split data into features and target
            X = self.data['Processed_Text']
            y = self.data['Satisfaction_Level']
            
            # Encode the target variable
            y_encoded = self.label_encoder.fit_transform(y)
            
            # Split into training and testing sets
            self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(
                X, y_encoded, test_size=self.test_size, random_state=self.random_state, stratify=y_encoded
            )
            
            print(f"Data prepared for modeling: {len(self.X_train)} training samples, {len(self.X_test)} testing samples")
            return True
            
        except Exception as e:
            print(f"Error preparing data for modeling: {e}")
            return False

    def train_behavior_model(self):
        """Train a behavior prediction model."""
        if self.X_train is None or self.y_train is None:
            print("Training data not prepared. Please prepare data first.")
            return None
            
        try:
            print("Training user behavior prediction model...")
            
            # Create a TF-IDF vectorizer + RandomForest pipeline
            pipeline = Pipeline([
                ('tfidf', TfidfVectorizer(max_features=5000, ngram_range=(1, 2))),
                ('classifier', MultinomialNB())  # Using MultinomialNB which works well with text features
            ])
            
            # Train the model
            self.behavior_model = pipeline.fit(self.X_train, self.y_train)
            
            # Evaluate the model
            y_pred = self.behavior_model.predict(self.X_test)
            accuracy = accuracy_score(self.y_test, y_pred)
            
            print(f"Model training complete. Test accuracy: {accuracy:.4f}")
            
            # Print classification report
            target_names = [str(self.label_encoder.inverse_transform([i])[0]) for i in range(len(self.label_encoder.classes_))]
            print("\nClassification Report:")
            print(classification_report(self.y_test, y_pred, target_names=target_names))
            
            return self.behavior_model
            
        except Exception as e:
            print(f"Error training behavior model: {e}")
            return None

    def identify_keywords(self, n_keywords=10):
        """Extract key terms influencing user satisfaction."""
        if self.behavior_model is None:
            print("Behavior model not trained. Please train the model first.")
            return {}
            
        try:
            # Get TF-IDF vectorizer from pipeline
            tfidf = self.behavior_model.named_steps['tfidf']
            classifier = self.behavior_model.named_steps['classifier']
            
            # Get feature names
            feature_names = tfidf.get_feature_names_out()
            
            # Extract feature importances from classifier
            importances = classifier.feature_importances_
            
            # Sort importances
            indices = np.argsort(importances)[::-1]
            
            # Get top keywords for each satisfaction level
            keywords_by_class = {}
            
            # For XGBoost with multiple classes
            class_keywords = {}
            for i, class_label in enumerate(self.label_encoder.classes_):
                # Get feature importance for this class
                importance_for_class = importances
                indices = np.argsort(importance_for_class)[::-1]
                
                # Get top keywords
                top_keywords = [(feature_names[idx], importance_for_class[idx]) for idx in indices[:n_keywords]]
                class_keywords[class_label] = top_keywords
            
            return class_keywords
            
        except Exception as e:
            print(f"Error identifying keywords: {e}")
            return {}

    def generate_improvement_suggestions(self, feedback_text):
        """Generate improvement suggestions based on user feedback."""
        # Basic cleaning of input
        if not isinstance(feedback_text, str) or feedback_text.strip() == "":
            return {
                'Sentiment': 'Neutral',
                'Predicted_Satisfaction': 'Unknown',
                'Improvement_Suggestions': ["Please provide more detailed feedback to receive meaningful suggestions."]
            }            
        try:
            # Preprocess the feedback
            processed_text = self.preprocess_text(feedback_text)
            
            # Analyze sentiment
            sentiment_result = self.analyze_sentiment_with_lexicon(feedback_text)
            sentiment = sentiment_result['label']
            sentiment_score = sentiment_result['score']
            
            # Predict satisfaction level
            if self.behavior_model:
                predicted_level_encoded = self.behavior_model.predict([processed_text])[0]
                predicted_satisfaction = self.label_encoder.inverse_transform([predicted_level_encoded])[0]
                print(f"Predicted Level (Encoded): {predicted_level_encoded}")
                print(f"Predicted Satisfaction: {predicted_satisfaction}")
            else:
                predicted_satisfaction = 'Unknown'
            
            # Generate improvement suggestions
            improvement_suggestions = []
            
            # Define keyword-based suggestion mappings (enhanced from original)
            suggestions_map = {
                'battery': [
                    "Consider increasing the battery capacity for longer usage.",
                    "Implement a low-power mode to extend battery life.",
                    "Add battery status indicators to improve user awareness."
                ],
                'grip': [
                    "Enhance the ergonomic design for a better grip.",
                    "Use non-slip materials to improve handling.",
                    "Consider adjustable grip sizes for different users."
                ],
                'customize': [
                    "Add more customizable settings tailored to user preferences.",
                    "Introduce presets for different types of food.",
                    "Develop a personalization system that learns from usage patterns."
                ],
                'customizable': [
                    "Add more customizable settings tailored to user preferences.",
                    "Introduce presets for different types of food.",
                    "Develop a personalization system that learns from usage patterns."
                ],
                'personalize': [
                    "Add more customizable settings tailored to user preferences.",
                    "Introduce presets for different types of food.",
                    "Develop a personalization system that learns from usage patterns."
                ],
                'quality': [
                    "Maintain the high-quality standards in materials and design.",
                    "Conduct regular quality checks to ensure product reliability.",
                    "Consider premium materials that enhance durability and user experience."
                ],
                'features': [
                    "Explore user-requested features like temperature control or smart alerts.",
                    "Consider integrating a companion app for enhanced functionality.",
                    "Implement AI-driven features that adapt to user behavior patterns."
                ],
                'app': [
                    "Improve the companion app interface for better usability.",
                    "Add more visualization features to track nutrition data.",
                    "Implement social sharing capabilities for recipes and food experiences."
                ],
                'color': [
                    "Introduce a variety of color options for personalization.",
                    "Offer limited edition colors based on user feedback.",
                    "Consider customizable RGB lighting elements for visual feedback."
                ],
                'nutrition': [
                    "Implement tracking features for nutritional intake.",
                    "Consider adding a recipe suggestion feature based on dietary needs.",
                    "Enhance accuracy of nutritional analysis algorithms."
                ],
                'price': [
                    "Evaluate pricing strategies to ensure competitiveness.",
                    "Consider offering bundle deals for multiple purchases.",
                    "Introduce a tiered product line with different price points."
                ],
                'salt': [
                    "Adjust the salt detection sensitivity based on user feedback.",
                    "Consider offering options for low-sodium alternatives.",
                    "Implement a feature for users to customize their preferred salt level detection."
                ],
                'temperature': [
                    "Add temperature sensing capabilities to provide food safety information.",
                    "Implement temperature-based cooking recommendations.",
                    "Consider alerts for foods that are too hot to consume safely."
                ],
                'weight': [
                    "Improve the accuracy of the weight sensing mechanism.",
                    "Add portion control features based on weight detection.",
                    "Implement personalized portion recommendations based on dietary goals."
                ],
                'connect': [
                    "Enhance Bluetooth/WiFi connectivity for more reliable data transfer.",
                    "Add seamless integration with other smart kitchen devices.",
                    "Improve cloud synchronization for multi-device users."
                ],
                'charge': [
                    "Implement wireless charging capabilities for convenience.",
                    "Improve charging efficiency to reduce downtime.",
                    "Add quick-charge functionality for users on the go."
                ],
                'clean': [
                    "Enhance water-resistant properties for easier cleaning.",
                    "Consider self-cleaning features for improved hygiene.",
                    "Redesign components for dishwasher-safe capabilities."
                ],
                'comfort': [
                    "Optimize weight distribution for better comfort during extended use.",
                    "Use softer materials at contact points for improved handling.",
                    "Consider ergonomic improvements based on diverse user testing."
                ],
                'data': [
                    "Improve data visualization for easier interpretation of nutritional information.",
                    "Implement enhanced data privacy controls for user information.",
                    "Add export functionality for integration with other health apps."
                ]
            }
            
            # Context-based suggestions based on sentiment and predicted satisfaction
            context_based_suggestions = {
                'Negative': {
                    'low': [
                        "Focus on core functionality improvements for better user satisfaction.",
                        "Consider a user experience redesign based on customer feedback.",
                        "Implement quality control measures to address potential issues."
                    ],
                    'medium': [
                        "Address specific user concerns while maintaining existing positive features.",
                        "Consider incremental improvements to gradually enhance user experience.",
                        "Develop a customer feedback program to better understand pain points."
                    ],
                    'high': [
                        "Though satisfaction is high, address negative feedback to maintain loyalty.",
                        "Consider minor refinements to address specific concerns.",
                        "Implement a proactive communication strategy to address potential issues."
                    ]
                },
                'Neutral': {
                    'low': [
                        "Focus on creating more emotional connection with the product experience.",
                        "Consider adding standout features to differentiate from competitors.",
                        "Implement user engagement strategies to increase product attachment."
                    ],
                    'medium': [
                        "Enhance unique selling points to convert neutral users to advocates.",
                        "Consider personalization features to increase user connection.",
                        "Develop community features to enhance product engagement."
                    ],
                    'high': [
                        "Build on existing strengths while adding innovative features.",
                        "Consider loyalty programs to maintain high satisfaction levels.",
                        "Implement user story collection to identify growth opportunities."
                    ]
                },
                'Positive': {
                    'low': [
                        "Focus on consistency to ensure all features meet positive expectations.",
                        "Address potential barriers to full product satisfaction.",
                        "Implement continuous improvement processes based on positive feedback."
                    ],
                    'medium': [
                        "Enhance features that receive the most positive feedback.",
                        "Consider expanding product line with complementary offerings.",
                        "Develop ambassador programs to leverage satisfied customers."
                    ],
                    'high': [
                        "Maintain high quality standards while exploring innovation opportunities.",
                        "Consider premium product variants for highly satisfied customers.",
                        "Implement case study collection to showcase positive outcomes."
                    ]
                }
            }
            
            # First add keyword-based suggestions
            for keyword, suggestions in suggestions_map.items():
                if keyword in feedback_text.lower():
                    for suggestion in suggestions:
                        if suggestion not in improvement_suggestions:
                            improvement_suggestions.append(suggestion)
            
            # Then add context-based suggestions
            # Determine satisfaction level category
            satisfaction_category = 'medium'  # default
            if predicted_satisfaction != 'Unknown':
                try:
                    satisfaction_value = float(predicted_satisfaction)
                    if satisfaction_value < 3:
                        satisfaction_category = 'low'
                    elif satisfaction_value > 4:
                        satisfaction_category = 'high'
                    else:
                        satisfaction_category = 'medium'
                except:
                    # If predicted_satisfaction is not numeric
                    if 'low' in predicted_satisfaction.lower() or '1' in predicted_satisfaction or '2' in predicted_satisfaction:
                        satisfaction_category = 'low'
                    elif 'high' in predicted_satisfaction.lower() or '4' in predicted_satisfaction or '5' in predicted_satisfaction:
                        satisfaction_category = 'high'
            
            # Add context-based suggestions
            if sentiment in context_based_suggestions and satisfaction_category in context_based_suggestions[sentiment]:
                context_suggestions = context_based_suggestions[sentiment][satisfaction_category]
                for suggestion in context_suggestions:
                    if suggestion not in improvement_suggestions:
                        improvement_suggestions.append(suggestion)
            
            # If no suggestions were found, add a default one
            if not improvement_suggestions:
                improvement_suggestions = [
                    "Thank you for your feedback! We continually monitor user experiences to improve our products.",
                    "Consider participating in our user research program to help shape future features.",
                    "We appreciate your input and will use it to enhance the Smart Spoon experience."
                ]
            
            # Limit to top 5 most relevant suggestions
            improvement_suggestions = improvement_suggestions[:5]
            
            return {
                'Sentiment': sentiment,
                'Sentiment_Score': sentiment_score,
                'Predicted_Satisfaction': predicted_satisfaction,
                'Improvement_Suggestions': improvement_suggestions
            }
            
        except Exception as e:
            print(f"Error generating improvement suggestions: {e}")
            return {
                'Sentiment': 'Error',
                'Predicted_Satisfaction': 'Error',
                'Improvement_Suggestions': ["An error occurred while processing your feedback. Please try again."]
            }

    def visualize_sentiment_distribution(self):
        """Visualize the sentiment distribution in the dataset."""
        if self.data is None or 'Sentiment' not in self.data.columns:
            print("Sentiment data not available. Please run sentiment analysis first.")
            return
            
        plt.figure(figsize=(10, 6))
        sns.countplot(x='Sentiment', data=self.data, palette='viridis')
        plt.title('Sentiment Distribution in User Feedback')
        plt.xlabel('Sentiment')
        plt.ylabel('Count')
        plt.tight_layout()
        plt.show()
    
    def visualize_satisfaction_vs_sentiment(self):
        """Visualize relationship between satisfaction levels and sentiment."""
        if self.data is None or 'Sentiment' not in self.data.columns or 'Satisfaction_Level' not in self.data.columns:
            print("Required data not available. Please run preprocessing first.")
            return
            
        plt.figure(figsize=(12, 8))
        
        # Convert satisfaction level to numeric if possible
        try:
            self.data['Satisfaction_Numeric'] = pd.to_numeric(self.data['Satisfaction_Level'], errors='coerce')
            
            # Create crosstab
            cross_tab = pd.crosstab(self.data['Satisfaction_Numeric'], self.data['Sentiment'])
            
            # Plot heatmap
            sns.heatmap(cross_tab, annot=True, cmap='YlGnBu', fmt='d', cbar=True)
            plt.title('Relationship Between Satisfaction Level and Sentiment')
            plt.xlabel('Sentiment')
            plt.ylabel('Satisfaction Level')
            
        except:
            # Alternative visualization if satisfaction levels are not numeric
            cross_tab = pd.crosstab(self.data['Satisfaction_Level'], self.data['Sentiment'])
            sns.heatmap(cross_tab, annot=True, cmap='YlGnBu', fmt='d', cbar=True)
            plt.title('Relationship Between Satisfaction Level and Sentiment')
            plt.xlabel('Sentiment')
            plt.ylabel('Satisfaction Level')
            
        plt.tight_layout()
        plt.show()
    
    def generate_top_words(self):
        """Generate top words for different sentiment categories."""
        if self.data is None or 'Sentiment' not in self.data.columns or 'Processed_Text' not in self.data.columns:
            print("Required data not available. Please run preprocessing first.")
            return
            
        sentiment_categories = self.data['Sentiment'].unique()
        
        for sentiment in sentiment_categories:
            # Filter text by sentiment
            texts = self.data[self.data['Sentiment'] == sentiment]['Processed_Text'].tolist()
            
            if not texts:
                print(f"No data for {sentiment}")
                continue
                
            # Count word frequencies
            word_freq = {}
            for text in texts:
                for word in text.split():
                    if len(word) > 2:  # Skip very short words
                        word_freq[word] = word_freq.get(word, 0) + 1
            
            # Sort by frequency
            sorted_words = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)
            
            # Display top 15 words
            print(f"\nTop words for {sentiment} feedback:")
            for word, count in sorted_words[:15]:
                print(f"  {word}: {count}")
            
        # Create a simple bar chart for visualization
        plt.figure(figsize=(10, 6))
        sns.countplot(x='Sentiment', data=self.data, palette='viridis')
        plt.title('Sentiment Distribution in User Feedback')
        plt.xlabel('Sentiment')
        plt.ylabel('Count')
        plt.tight_layout()
        plt.show()
    
    def run_full_analysis(self):
        """Run the complete analysis pipeline."""
        print("Starting Smart Spoon Feedback Analysis...")
        
        # Step 1: Load data
        if not self.load_data():
            return "Analysis failed during data loading."
            
        # Step 2: Perform sentiment analysis
        self.perform_sentiment_analysis()
        
        # Step 3: Prepare data for modeling
        if not self.prepare_data_for_modeling():
            return "Analysis failed during data preparation."
            
        # Step 4: Train behavior model
        if not self.train_behavior_model():
            return "Analysis failed during model training."
            
        # Step 5: Identify keywords (simpler approach)
        print("\nAnalyzing key terms in feedback...")
        self.generate_top_words()
        
        print("\nAnalysis complete!")
        print("The system is now ready to analyze new feedback and generate personalized improvement suggestions.")
        
        return "Analysis completed successfully."

def analyze_user_feedback(feedback_text, analyzer):
    """Analyze a single user feedback and return results."""
    if analyzer is None:
        return {
            'Sentiment': 'Error',
            'Predicted_Satisfaction': 'Error',
            'Improvement_Suggestions': ["Analyzer not initialized. Please run the full analysis first."]
        }
        
    results = analyzer.generate_improvement_suggestions(feedback_text)
    
    # Print results
    print("\nAnalysis Results:")
    print(f"Sentiment: {results['Sentiment']} (Score: {results['Sentiment_Score']:.2f})")
    print(f"Predicted Satisfaction Level: {results['Predicted_Satisfaction']}")
    print("Improvement Suggestions:")
    for i, suggestion in enumerate(results['Improvement_Suggestions'], 1):
        print(f"{i}. {suggestion}")
        
    return results

# Example usage
def main():
    try:
        print("Smart Spoon Feedback Analyzer")
        print("============================")
        
        # Initialize analyzer
        analyzer = SmartSpoonFeedbackAnalyzer('response_data.csv')
        
        # Run full analysis
        analyzer.run_full_analysis()
        
        # Visualizations (simpler versions)
        # analyzer.visualize_sentiment_distribution()
        analyzer.visualize_satisfaction_vs_sentiment()
        
        # Interactive user input loop
        while True:
            user_review = input("\nPlease enter your feedback about the Smart Spoon (or type 'exit' to quit): ")
            
            if user_review.lower() == 'exit':
                print("Thank you for using the Smart Spoon Feedback Analyzer!")
                break
                
            analyze_user_feedback(user_review, analyzer)
            
    except KeyboardInterrupt:
        print("\nAnalysis terminated by user.")
    except Exception as e:
        print(f"\nAn unexpected error occurred: {e}")

if __name__ == "__main__":
    main()
```
